Logistic Regresyon
Bu veri setinde, belirli özelliklerin kadın çalışan oranıyla nasıl ilişkilendiğini anlamak için logistic regresyonu kullanıyoruz. Örneğin, belirli bir şirketin kadın çalışan oranının %50'den büyük olup olmadığını tahmin etmeye çalışıyoruz.

```python
import pandas as pd

# CSV dosyasını yükleme
df = pd.read_csv("C:\\Users\\user\\Downloads\\Diversity in tech companies.csv")
  # Dosya adını doğru yazdığından emin ol

# Verinin ilk birkaç satırını görüntüleme
print(df.head())

# Eksik ve hatalı değerleri temizleme
df.replace({'-': 0, '<1': 0.5}, inplace=True)
df = df.apply(pd.to_numeric, errors='coerce')  # Hatalı değerleri NaN yapar
df.dropna(inplace=True)  # NaN değerlerini kaldırır


print(df.head())  # Temizlenen veriyi kontrol etme
print(df.info())  # Veride eksik değer kalıp kalmadığını kontrol etme

# Özellikler (X) ve hedef (y) belirleme
X = df[['Male %', '% White', '% Asian', '% Latino', '% Black', '% Multi', '% Other', '% Undeclared']]
y = df['Female %'] > 50  # Örnek bir eşik belirledik
print(X.head())  # Özelliklerimizi kontrol etme
print(y.head())  # Hedef değişkenimizi kontrol etme

from sklearn.model_selection import train_test_split

# Eğitim ve test setlerine ayırma
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

print(f"Training set size: {len(X_train)}, Test set size: {len(X_test)}")

from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix

# Verileri ölçeklendirme
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Modeli oluşturma ve eğitim
model = LogisticRegression()
model.fit(X_train, y_train)

# Tahmin yapma
y_pred = model.predict(X_test)

# Model değerlendirmesi
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

Korelasyon Analizi
Belirli değişkenler arasındaki ilişkilerin yönünü (pozitif veya negatif) ve gücünü belirlememizi sağlar. Örneğin, "Female %" ile diğer demografik değişkenler arasında bir ilişki olup olmadığını görebiliriz.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# CSV dosyasını yükleme
df = pd.read_csv("C:\\Users\\user\\Downloads\\Diversity in tech companies.csv")


# Eksik ve hatalı değerleri temizleme
df.replace({'-': 0, '<1': 0.5}, inplace=True)
df = df.apply(pd.to_numeric, errors='coerce')  # Hatalı değerleri NaN yapar
df.dropna(inplace=True)  # NaN değerlerini kaldırır
df = df.fillna(df.mean())  # NaN değerlerini sütun ortalamaları ile doldurma

# Sadece sayısal sütunları seçin
df_numeric = df.select_dtypes(include=['float64', 'int64'])

# NaN değerleri içeren satırları atma
df_numeric = df_numeric.dropna()

# Korelasyon matrisini hesaplama
corr_matrix = df_numeric.corr()

# Korelasyon matrisini yazdırma
print(corr_matrix)

# Isı haritası ile korelasyon matrisini görselleştirme
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Korelasyon Matrisi Isı Haritası')
plt.show()
```

```python
Female % ile diğer demografik değişkenler(cinsiyet,yaş,etnik köken) arasındaki ilişkiyi incelemek için korelasyon analizi
Year ve % White ile zayıf pozitif korelasyon,
Male % ile  negatif korelasyon var.
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# CSV dosyasını yükleme
df = pd.read_csv("C:\\Users\\user\\Downloads\\Diversity in tech companies.csv")
# NaN değerleri kontrol et
df = df.dropna()  # Tüm NaN değerleri içeren satırları atma
# Sadece sayısal sütunları seçme
df_numeric = df.select_dtypes(include=['float64', 'int64'])

# Korelasyon matrisini hesapla
corr_matrix = df_numeric.corr()

# "Female %" ile diğer değişkenler arasındaki korelasyonu yazdır
print(corr_matrix['Female %'])
# Scatter plot oluşturma
demographic_columns = ['% White', '% Asian', '% Latino', '% Black', '% Multi', '% Other', '% Undeclared']

for column in demographic_columns:
    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=df, x=column, y='Female %')
    plt.title(f'Female % ile {column} Arasındaki İlişki')
    plt.xlabel(column)
    plt.ylabel('Female %')
    plt.axhline(y=df['Female %'].mean(), color='r', linestyle='--', label='Ortalama Female %')
    plt.legend()
    plt.show()

```
